### В файле содержится пример использования класса Crowdhuman2coco.
Указываем пути:
- до папки с картинками иходного датасета;
- до папки, куда сохранятся картинки (в данном примере для тестового dataset);
- до файла json с сохраненным описанием для нового dataset;
- начальную позицию, с которой начнется запись;
- число записываемых картинок.

После запуска кода из папки val_images первые 2000 картинок будут пропущены, начиная с 2001 картинки остальные в количестве 1500 штук будут записаны в папку test.\
Для обучаещего датасета бюли использованы первые 7500 картинок из train_images, для валидации использовал первые 1500 первых картинок из val_images. Для тестового набора использовал 1500 картинок, начиная с 2001 из папки val_images.


```python
from crowdhuman2coco import Crowdhuman2coco


convert = Crowdhuman2coco(path_to_annotation='data\\annotation_val.odgt',
                    path_to_data='data/val_images\\Images',
                    path_to_new_data='data\\test',
                    path_to_json='data/annotation_test.json',
                    im_start= 2000,
                    number_images=1500)
convert.convert2coco(args_head=1, args_rm_hocc=0, args_rm_hunsure=1,
             args_rm_hignore=0)
```
